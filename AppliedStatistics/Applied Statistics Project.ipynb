# 2. An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the 
failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following 
questions. [4 Marks]
2.A. Probability that none of the items are defective? [1 Mark]
2.B. Probability that exactly one of the items is defective? [1 Mark]
2.C. Probability that two or fewer of the items are defective? [1 Mark]
2.D. Probability that three or more of the items are defective ? [1 Mark]
We will use appropriate Python libraries to calculate the probabilities for each of the scenarios described in your question regarding the electrical manufacturing company. The binomial probability distribution will be used, as it is appropriate for this scenario involving a fixed number of trials (10 items), two possible outcomes (defective or not defective), and a constant probability of success (failure rate of 5% or 0.05).

This code utilizes the scipy.stats module's binom class to calculate the probabilities. The .pmf method is used for calculating the probability mass function for a specific number of defects, and .cdf is used for the cumulative distribution function to calculate the probability of two or fewer defects. The complement of this cumulative probability gives the probability of three or more defects.

```python
# from scipy.stats import binom

# Define the parameters
n = 10  # number of items in the sample
p = 0.05  # probability of an item being defective

# Probability calculations
# 2.A. Probability that none of the items are defective
prob_none_defective = binom.pmf(0, n, p)

# 2.B. Probability that exactly one of the items is defective
prob_one_defective = binom.pmf(1, n, p)

# 2.C. Probability that two or fewer of the items are defective
prob_two_or_fewer_defective = binom.cdf(2, n, p)

# 2.D. Probability that three or more of the items are defective
prob_three_or_more_defective = 1 - prob_two_or_fewer_defective

# Print the results
print(f"Probability that none of the items are defective: {prob_none_defective:.4f}")
print(f"Probability that exactly one of the items is defective: {prob_one_defective:.4f}")
print(f"Probability that two or fewer of the items are defective: {prob_two_or_fewer_defective:.4f}")
print(f"Probability that three or more of the items are defective: {prob_three_or_more_defective:.4f}")
```

# 3. A car salesman sells on an average 3 cars per week. [3 Marks]
3.A. What is Probability that in a given week he will sell some cars? [1 Mark]
3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars? [1 Mark]
3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week. [1 Mark]
we will use the Poisson distribution, which is a discrete probability distribution that expresses the probability of a given number of events happening in a fixed interval of time or space if these events occur with a known constant mean rate and independently of the time since the last event. In this case, the events are the sale of cars, and the mean rate is 3 cars per week.

Let's calculate each part using Python libraries:

3.A. The probability that in a given week he will sell some cars is the complement of the probability of selling no cars.

3.B. The probability that in a given week he will sell 2 or more but less than 5 cars can be calculated by taking the cumulative probability of selling 4 cars and subtracting the cumulative probability of selling 1 car.

3.C. We will plot the Poisson distribution function for the cumulative probability of cars sold per week versus the number of cars sold per week.

This code will calculate the required probabilities and generate a plot that visualizes the cumulative probability of the number of cars sold per week according to a Poisson distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Average number of cars sold per week
lambda_ = 3  # 'lambda' is the average rate (3 cars/week)

# 3.A. Probability of selling some cars (1 or more)
prob_some_cars = 1 - poisson.pmf(0, lambda_)

# 3.B. Probability of selling 2 or more but less than 5 cars
# P(2 ≤ X < 5) = P(X ≤ 4) - P(X ≤ 1)
prob_2_to_4_cars = poisson.cdf(4, lambda_) - poisson.cdf(1, lambda_)

# Print the results for 3.A and 3.B
print(f"Probability of selling some cars (1 or more): {prob_some_cars:.4f}")
print(f"Probability of selling 2 or more but less than 5 cars: {prob_2_to_4_cars:.4f}")

# 3.C. Plot the Poisson distribution function for cumulative probability
x = np.arange(0, 11)  # Define the range of cars sold (0 to 10)
y = poisson.cdf(x, lambda_)  # Cumulative probabilities

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o')
plt.title('Poisson Distribution - Cumulative Probability of Cars Sold Per Week')
plt.xlabel('Number of Cars Sold')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()
```

# 4. Accuracy in understanding orders for a speech based bot at a restaurant is important for the Company X which has designed, 
marketed and launched the product for a contactless delivery due to the COVID-19 pandemic. Recognition accuracy that measures the 
percentage of orders that are taken correctly is 86.8%. Suppose that you place an order with the bot and two friends of yours 
independently place orders with the same bot. Answer the following questions. [3 Marks] 
4.A. What is the probability that all three orders will be recognised correctly? [1 Mark]
4.B. What is the probability that none of the three orders will be recognised correctly? [1 Mark]
4.C. What is the probability that at least two of the three orders will be recognised correctly? [1 Mark]
we can use the binomial probability distribution since we have a series of independent and identical trials (placing orders with the bot) with two possible outcomes (correctly recognized or not). The probability of success (correct recognition) is given as 86.8% or 0.868.

This code uses the binom.pmf method from the scipy.stats module to calculate the probability mass function for the given number of successes (correctly recognized orders) out of the total number of trials (orders placed). It calculates the probabilities for each of the scenarios you mentioned.

```python
from scipy.stats import binom

# Define the parameters for the binomial distribution
n = 3  # number of orders
p = 0.868  # probability of an order being recognized correctly

# 4.A. Probability that all three orders will be recognized correctly
prob_all_correct = binom.pmf(3, n, p)

# 4.B. Probability that none of the three orders will be recognized correctly
prob_none_correct = binom.pmf(0, n, p)

# 4.C. Probability that at least two of the three orders will be recognized correctly
# This is the sum of the probabilities of exactly two and exactly three orders being correct
prob_at_least_two_correct = binom.pmf(2, n, p) + binom.pmf(3, n, p)

# Print the results
print(f"Probability that all three orders will be recognized correctly: {prob_all_correct:.4f}")
print(f"Probability that none of the three orders will be recognized correctly: {prob_none_correct:.4f}")
print(f"Probability that at least two of the three orders will be recognized correctly: {prob_at_least_two_correct:.4f}")
```

Each step in the script is targeted toward extracting meaningful insights from the dataset, such as understanding the funding patterns, the impact of startup performance in competitions on their operational status, and specific trends in startup events. This type of analysis is valuable for organizations focusing on the startup ecosystem, as it can inform investment decisions, event focus, and market analysis.







```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# Q1 - Read the CSV file
file_path = '/Users/snbaskarraj/Documents/UT/AppliedStatistics/DS+-+Part3+-+CompanyX_EU.csv'
df = pd.read_csv(file_path)

# Q2 - Data Exploration
# A. Check datatypes
print("Data Types:\n", df.dtypes)

# B. Check for null values
print("\nNull Values:\n", df.isnull().sum())

# Q3 - Data preprocessing & visualization
# A. Drop null values
df.dropna(inplace=True)

# B. Convert 'Funding' to numerical value
df['Funds_in_million'] = df['Funding'].apply(lambda x: float(x[1:-1])/1000 if x[-1] == 'K' 
                                             else (float(x[1:-1])*1000 if x[-1] == 'B' else float(x[1:-1])))

# C. Plot boxplot for funds in million
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Funds_in_million')
plt.title('Boxplot of Funds in Million')
plt.show()

# D. Number of outliers greater than the upper fence
Q1 = df['Funds_in_million'].quantile(0.25)
Q3 = df['Funds_in_million'].quantile(0.75)
IQR = Q3 - Q1
upper_fence = Q3 + 1.5 * IQR
outliers = df[df['Funds_in_million'] > upper_fence]
print("Number of Outliers:", outliers.shape[0])

# E. Frequency of OperatingState feature classes
print("\nFrequency of OperatingState:\n", df['OperatingState'].value_counts())

# Q4 - Statistical Analysis
# A, B & C - Significance test
operating = df[df['OperatingState'] == 'Operating']['Funds_in_million']
closed = df[df['OperatingState'] == 'Closed']['Funds_in_million']

# Null hypothesis: There is no significant difference in funds raised between operating and closed companies.
# Alternative hypothesis: There is a significant difference in funds raised between operating and closed companies.
t_stat, p_value = ttest_ind(operating, closed, equal_var=False)
print("\nT-test results:\nT-statistic:", t_stat, "\nP-value:", p_value)

# D - Make a copy of the original data frame
df_copy = df.copy()

# E - Frequency distribution of Result variables
print("\nFrequency Distribution of Result:\n", df_copy['Result'].value_counts())

# F - Calculate percentage of winners and contestants that are still operating
winners_operating = df_copy[(df_copy['Result'] == 'Winner') & (df_copy['OperatingState'] == 'Operating')].shape[0]
total_winners = df_copy[df_copy['Result'] == 'Winner'].shape[0]
winners_operating_percent = (winners_operating / total_winners) * 100

contestants_operating = df_copy[(df_copy['Result'] == 'Contestant') & (df_copy['OperatingState'] == 'Operating')].shape[0]
total_contestants = df_copy[df_copy['Result'] == 'Contestant'].shape[0]
contestants_operating_percent = (contestants_operating / total_contestants) * 100

print("\nPercentage of Winners Still Operating:", winners_operating_percent)
print("Percentage of Contestants Still Operating:", contestants_operating_percent)

# G, H - Hypothesis testing between winners and contestants
# Null hypothesis: There is no difference in proportions of operating companies between winners and contestants.
# Alternative hypothesis: There is a difference in proportions of operating companies between winners and contestants.
# This will be a proportion z-test, which is not directly available in scipy, thus manually calculated or use statsmodels

# I - Select events with 'disrupt' keyword from 2013 onwards
df_disrupt = df_copy[df_copy['Event'].str.contains('disrupt') & 
                     (df_copy['Event'].str.contains('2013') |
                      df_copy['Event'].str.contains('2014') |
                      df_copy['Event'].str.contains('2015') |
                      df_copy['Event'].str.contains('2016') |
                      df_copy['Event'].str.contains('2017') |
                      df_copy['Event'].str.contains('2018') |
                      df_copy['Event'].str.contains('2019') |
                      df_copy['Event'].str.contains('2020') |
                      df_copy['Event'].str.contains('2021') |
                      df_copy['Event'].str.contains('2022'))]

print("Selected Disrupt Events from 2013 onwards:\n", df_disrupt['Event'])
```
